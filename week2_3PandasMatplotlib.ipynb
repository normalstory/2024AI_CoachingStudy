**Boston House Price Dataset**
- 머신 러닝 등 데이터 분석을 처음 배울 때, 가장 대표적으로 사용하는 Example Dataset
- 1978년에 발표된 데이터로, 미국 인구통계 조사 결과 미국 보스턴 지역의 주택 가격에 영향 요소들을 정리함
- http://lib.stat.cmu.edu/datasets/boston 


[01]  CRIM
자치시(town) 별 1인당 범죄율

 [02]  ZN
25,000 평방피트를 초과하는 거주지역의 비율

 [03]  INDUS
비소매상업지역이 점유하고 있는 토지의 비율

 [04]  CHAS
찰스강에 대한 더미변수(강의 경계에 위치한 경우는 1, 아니면 0)

 [05]  NOX 
10ppm 당 농축 일산화질소

 [06]  RM
주택 1가구당 평균 방의 개수 

 [07]  AGE 
1940년 이전에 건축된 소유주택의 비율 

 [08]  DIS
5개의 보스턴 직업센터까지의 접근성 지수 

 [09]  RAD
방사형 도로까지의 접근성 지수 

 [10]  TAX
10,000 달러 당 재산세율 

 [11]  PTRATIO
자치시(town)별 학생/교사 비율

 [12]  B
1000(Bk-0.63)^2, 여기서 Bk는 자치시별 흑인의 비율을 말함. 

 [13]  LSTAT
모집단의 하위계층의 비율(%)

 [14]  MEDV
본인 소유의 주택가격(중앙값) (단위: $1,000)

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data' #Data URL
data_url = 'data/housing.data'
df_data = pd.read_csv(data_url, sep='\s+', header = None) #csv 타입 데이터 로드, separate는 빈공간으로 지정하고, Column은 없음

df_data.columns = [ # Column Header 이름 지정 
    'CRIM','ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO' ,'B', 'LSTAT', 'MEDV'] 

df_data.head()
```

1. 상관관계 분석 - scatter graph 

```python
fig = plt.figure()
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))
    
ax[0].scatter(df_data["CRIM"], df_data["MEDV"]) # 범죄율 - 집값 
ax[1].scatter(df_data["PTRATIO"], df_data["MEDV"])
ax[2].scatter(df_data["AGE"], df_data["MEDV"])
ax[3].scatter(df_data["NOX"], df_data["MEDV"])
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10.0, 10.0, forward=True)
ax = []
for i in range(1,5):
    ax.append(fig.add_subplot(2,2,i))

columns = ["CRIM", "PTRATIO", "AGE", "NOX"]
colors = ["b", "g", "c", "r"]
for i in range(4):
    ax[i].scatter(df_data[columns[i]], df_data["MEDV"], color=colors[i], label=columns[i])
    ax[i].legend()
    ax[i].set_title(columns[i])

plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

- plot 함수를 통해 전체 데이터 각각의 graph 파악 

```python
df_data.plot()
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(5,5)
ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)
ax_1.plot(df_data["MEDV"])
ax_2.hist(df_data["MEDV"], bins=50)
ax_1.set_title("House price MEDV")
ax_2.set_title("House price MEDV")

plt.show() # plot으로는 잘 안보이지만 hist로 보면 명확한 정규분포를 확인할 수 있다 
```

- Scaled boxplot - 아웃라이어 체크 

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scale_data = std_scaler.fit_transform(df_data)
scale_data
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.boxplot(scale_data, labels=df_data.columns) # 아웃라이어가 너무 많으면 데이터 자체의 의미를 신뢰하기 어렵기 때문에 boxplot으로 체크 
fig.set_size_inches(10, 10)
plt.show() 
```

- Scatter matrix - 전체적인 상관관계 확인 

```python
from pandas.plotting import scatter_matrix
pd.plotting.scatter_matrix(df_data, diagonal="kde", alpha=1,  figsize=(20, 20))
plt.show()
```

- Scatter matrix - co-relationship check

```python
corr_data = np.corrcoef(scale_data.T[:5])
corr_data
```

```python
corr_data.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr_data, vmin=-1, vmax=1,  interpolation='nearest')
fig.colorbar(cax)
fig.set_size_inches(10,10)
ticks= np.arange(0,5,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(df_data.columns[:5])
ax.set_yticklabels(df_data.columns[:5])

plt.show() # 색이 진할수록 상관관계 있다  
```

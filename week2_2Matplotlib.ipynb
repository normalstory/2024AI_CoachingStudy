# **matplotlib**

- pyplot 객체를 사용하여 데이터를 표시 
    - Graph는 원래 figure(캔버스) 객체에 생성됨
        - plot 사용시, 
            - pyplot 객체 내 - 기본 figure에 그래프가 그려짐
            - plot 추가될 때마다 figure(캔버스) 객체에 다른 색으로 그래프들을 쌓음 
        - show 로 
            - 화면에 flush(그대로 출력, 캔버스 초기화)함 


최대 단점 
- argument를 kwargs 받음
- 고정된 argument가 없어서 alt+tab으로 확인이 어려움

```python
import matplotlib.pyplot as plt
import numpy as np 
import matplotlib.pyplot as plt
```

1. plot

```python
X = range(100)
Y = [value**2 for value in X]
plt.plot (X, Y) # X, Y 값은 list 타입 
plt.show()
```

```python
X = range(100)
Y = range(100)
plt.plot(X, Y)
plt.show()
```

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.plot(Y_2, X_1)
plt.plot(range(100), range(100))
plt.show()
```

2. add_subplot

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10,10) # 크기지정
ax_1 = fig.add_subplot(1,2,1) # 두개의 plot 생성
ax_2 = fig.add_subplot(1,2,2)  # 두개의 plot 생성

ax_1.plot(X_1, Y_1, c="b")  # 첫번째 plot
ax_2.plot(X_2, Y_2, c="g")  # 두번째 plot
plt.show() # show & flush
```

6. Bar chart

```python
# Y 축의 값 
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]
# X 축의 값 : 0,1,2,.. 
X = np.arange(4)

plt.bar(X + 0.00, data[0], color = 'b', width = 0.25) # X축(X + 0.00) , Y축(data[0]), width(넓이) 
plt.bar(X + 0.25, data[1], color = 'g', width = 0.25)
plt.bar(X + 0.50, data[2], color = 'r', width = 0.25)
plt.xticks(X + 0.25, ("A","B","C","D"))
plt.show()
```

7. barh

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b')
plt.show()
```
